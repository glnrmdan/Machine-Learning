```python
# Convert waktu training ke dalam menit
time_training = ['12 menit 36 detik', '31 menit']

# Hitung selisih waktu training terbesar dan terkecil
time_training_seconds = [12*60+36, 31*60]
time_training_seconds = np.array(time_training_seconds)
max_time_training = np.max(time_training_seconds)
min_time_training = np.min(time_training_seconds)
diff_time_training = max_time_training - min_time_training
print("diff seconds: ", diff_time_training)

# kemudian selisih tersebut buat persentasenya
time_training_percentage = (diff_time_training / min_time_training) * 100
time_training_percentage = round(time_training_percentage, 2)
time_training_percentage
```

```python
def convert_to_minutes(time_str):
    parts = time_str.split()
    if len(parts) == 4:  # format: xx menit yy detik
        minutes = int(parts[0])
        seconds = int(parts[2])
        return minutes + seconds / 60
    elif len(parts) == 2:  # format: xx menit
        return int(parts[0])
    else:
        return 0

time_minutes = [convert_to_minutes(t) for t in time_training]
```

```python
# Plotting metrik evaluasi
fig, axs = plt.subplots(2, 1, figsize=(16, 10))

metrics = ['Presisi', 'Recall', 'Akurasi', 'F1-score', 'IoU']
metric_data = [precision, recall, accuracy, f1_score, iou]

for i, metric in enumerate(metrics):
    ax = axs[0]
    ax.bar(np.arange(len(scenarios)) + 0.1 * i, metric_data[i], width=0.1, label=metric)

ax.set_xticks(np.arange(len(scenarios)))
ax.set_xticklabels(scenarios, fontsize=16)
ax.set_yticklabels([f'{t:.2f}' for t in ax.get_yticks()], fontsize=14)
ax.set_ylabel('Nilai', fontsize=16)
ax.set_xlabel('Skenario', fontsize=16)
ax.set_title('Metrik Evaluasi', fontsize=20)
# Legend location under the plot
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.16), ncol=5, fontsize=16)
ax.grid(color='gray', linestyle='--', linewidth=0.5)

# Plotting durasi waktu training menggunakan bar plot
ax = axs[1]
ax.plot(scenarios, time_minutes, marker='o', linestyle='-', color='b')
ax.set_xticklabels(scenarios, rotation=45, fontsize=16)
ax.set_yticklabels([f'{t:.0f}' for t in ax.get_yticks()], fontsize=14)
ax.set_xlabel('Skenario', fontsize=16)
ax.set_ylabel('Waktu Training (menit)', fontsize=16)
ax.set_title('Durasi Waktu Training', fontsize=20)
ax.grid(color='gray', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data from the previous calculations
total_pixels_gt = 41385984
total_vegetation_count_gt = 17036636 
total_non_vegetation_count_gt = 24349348 
vegetation_percentage_gt = (total_vegetation_count_gt / total_pixels_gt) * 100
non_vegetation_percentage_gt = (total_non_vegetation_count_gt / total_pixels_gt) * 100

total_pixels_pred = 41385984
total_vegetation_count_pred = 9793303 
total_non_vegetation_count_pred = 31592681 
vegetation_percentage_pred = (total_vegetation_count_pred / total_pixels_pred) * 100
non_vegetation_percentage_pred = (total_non_vegetation_count_pred / total_pixels_pred) * 100

# Print the total results
print("Ground Truth:")
print("Total Pixels:", total_pixels_gt)
print(f"Total vegetation pixels: {total_vegetation_count_gt} ({vegetation_percentage_gt:.2f}%)")
print(f"Total non-vegetation pixels: {total_non_vegetation_count_gt} ({non_vegetation_percentage_gt:.2f}%)")

print("Predicted Image:")
print("Total Pixels:", total_pixels_pred)
print(f"Total vegetation pixels: {total_vegetation_count_pred} ({vegetation_percentage_pred:.2f}%)")
print(f"Total non-vegetation pixels: {total_non_vegetation_count_pred} ({non_vegetation_percentage_pred:.2f}%)")

# Visualization
categories = ['Vegetasi', 'Non-Vegetasi']
ground_truth_counts = [vegetation_percentage_gt, non_vegetation_percentage_gt]
predicted_counts = [vegetation_percentage_pred, non_vegetation_percentage_pred]

x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, ground_truth_counts, width, label='Ground Truth')
bars2 = ax.bar(x + width/2, predicted_counts, width, label='Predicted')

ax.set_xlabel('Kateogori Piksel')
ax.set_ylabel('Persentase (%)')
ax.set_title('Perbandingan Distribusi Piksel Vegetasi dan Non-Vegetasi')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Add labels inside of the bars by percentage value and numerical value
for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height/2, f'{height:.2f}%', ha='center', va='center', color='white', fontsize=12)
    
    # Add numerical value inside of the bars
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{total_pixels_gt * height / 100:.0f}', ha='center', va='bottom', fontsize=12)
    
for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height/2, f'{height:.2f}%', ha='center', va='center', color='black', fontsize=12)
    
    # Add numerical value inside of the bars
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{total_pixels_pred * height / 100:.0f}', ha='center', va='bottom', fontsize=12)
    
fig.tight_layout()
plt.show()
```
